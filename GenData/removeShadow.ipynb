```python
def getRemovedShadowImg(img):
    arr = np.array(img)
    diff = arr - shadow_rgb
    diff = np.abs(diff)
    diff = np.sum(diff,axis=2)
    diff_idx = (diff<=40)
    diff_idx[:145:,::] = False
    arr[diff_idx] = [0,0,0,0]
    new = Image.fromarray(arr)
    return new

def import_crewpool():
    # assign directory
    directory = 'crewpool'
    img_list = []
    # iterate over files in
    # that directory
    for filename in os.listdir(directory):
        f = os.path.join(directory, filename)
        # checking if it is a file
        if os.path.isfile(f) and str(f)[-4::] in [".png",".jpg",".jpeg"]:
            img = Image.open(f)
            img_list.append(img)
    return img_list

img_with_shadow = import_crewpool()
n=0
for i in range(len(img_with_shadow)):
    img = img_with_shadow[i]
    new = getRemovedShadowImg(img)
    n+=1
    new.save("crewpool_noshadow/img"+str(n)+".png")
    print(n,"done")
print("DONE")

```
